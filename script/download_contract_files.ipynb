```python
from etherscan.client import BadRequest, ConnectionRefused, EmptyResponse
from etherscan.contracts import Contract
import os
import backoff
import json

ETHERSCAN_API = os.getenv('ETHERSCAN_API', '')

@backoff.on_exception(backoff.expo,
                      (EmptyResponse, BadRequest, ConnectionRefused),
                      max_tries=8)
def download_contract(address):
    print("Downloading contract: " + address)
    api = Contract(address=address, api_key=ETHERSCAN_API)
    sourcecode = api.get_sourcecode()
    return sourcecode

```

```python
address = '0x000000e92d78d90000007f0082006fda09bd5f11'
sourcecode = download_contract(address)
```
